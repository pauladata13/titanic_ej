La funcion Concat() es √∫til para apilar o unir datos, manteniendo o ajustando √≠ndices y etiquetas seg√∫n convenga, y se suele usar para combinar informaci√≥n en DataFrames o Series de pandas de forma flexible y eficiente.

```python
a = df.head(2)   
b = df.tail(2)   

resultado = pd.concat([a, b])

print(resultado)
```

```python
df_datos = df[['PassengerId', 'Name', 'Sex', 'Age']]
df_pagos = df[['PassengerId', 'Fare', 'Embarked']]

df_unido = pd.concat([df_datos, df_pagos[['Fare', 'Embarked']]], axis=1)

print(df_unido.head())
```


## UTILIDAD

El m√©todo merge() en pandas sirve para combinar dos DataFrames en base a columnas o √≠ndices comunes.
Devuelve un nuevo DataFrame que combina las columnas seleccionadas de df1 y df2, unidas por la columna com√∫n PassengerId.




```python
df1 = df[["PassengerId", "Name", "Sex"]]
df2 = df[["PassengerId", "Age", "Fare"]]
merged = df1.merge(df2, on="PassengerId")
merged.head()
```
